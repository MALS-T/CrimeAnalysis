```python
# Testing to find cause of issue in choropleth (personal reference)
import json
import pandas as pd
import plotly.express as px

geojson_testpath = 'testmapuniqueid.geojson'

with open(geojson_testpath, 'r') as f:
    geojson_testdata = json.load(f)

test_df = pd.DataFrame({
    'ID': ['a', 'b1', 'b2', 'c', 'd1', 'd2'],
    'Value': [1, 2, 2, 3, 4, 4],
    'HoverName': ['a', 'b', 'b', 'c', 'd', 'd']
})

for feature in geojson_testdata['features']:
    print(feature)

testchoropleth = px.choropleth(
    data_frame=test_df,
    geojson=geojson_testpath,
    locations='ID',
    featureidkey='properties.id',
    hover_name='HoverName',
    hover_data={'HoverName': False, 'Value': True, 'ID': False},
    color='Value',
    color_continuous_scale='Viridis',
    title='test map'
)

testchoropleth.update_geos(
    fitbounds='locations',     # Zoom to GeoJSON regions
    visible=False              # Hide base geography
)

testchoropleth.show()
```
